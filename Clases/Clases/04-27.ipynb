## Sigmoide

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.figure()
plt.plot(x, y)
plt.title("Función Sigmoide")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.show()
```
